### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pymoli = pd.read_csv(file_to_load)
pymoli.head()
```

## Player Count

* Display the total number of players


```python
total_players = pymoli['SN'].nunique()
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uni_items = pymoli['Item ID'].nunique()
avg_price = round(pymoli['Price'].mean(),2)
num_purchases = len(pymoli['Purchase ID'])
total_revenue = pymoli['Price'].sum()
Purchasing_Analysis = pd.DataFrame([{
                                    "Number of Unique Items" : uni_items,
                                    'Averge Price' : avg_price,
                                    'Number of Purchases' : num_purchases,
                                    'Total Revenue' : total_revenue
                                    }])
Purchasing_Analysis.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# I was struggling to figure this out late at night and I'm pretty sure this workflow is way more complex than it needs to be
# It's also, probably, really inefficient but I got it to work and I didn't have time to fix it
# A few exercises down it starts to look more normal when I started to get it more

dem = pymoli.groupby(['SN','Gender'])
demgrph = pd.DataFrame(dem.size())
demgrph = demgrph.reset_index()
demgrph
male = len(demgrph.loc[demgrph['Gender'] == 'Male'])
female = len(demgrph.loc[demgrph['Gender'] == 'Female'])
other = len(demgrph.loc[demgrph['Gender'] == 'Other / Non-Disclosed'])
#call in total from above
male_prct = round((male / total_players)*100,3)
female_prct = round((female / total_players)*100,3)
other_prct = round((other / total_players)*100,3)
genders = ["Male","Female","Other / Non-Disclosed"]
g_counts =[male,female,other]
g_prcnts = [male_prct,female_prct,other_prct]
Gender_Demographics_Analysis = pd.DataFrame(
                                        {'Gender':genders,
                                        'Total Count':g_counts,
                                        "Percent of Players" : g_prcnts})
Gender_Demographics_Analysis = Gender_Demographics_Analysis.set_index('Gender')
Gender_Demographics_Analysis
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
dem1 = pymoli.groupby(['Gender'])
t_p_v = dem1['Price'].sum()
p_cnt = dem1['Price'].count()
av_p_p = round(dem1['Price'].mean(),2)
#call in number of players from before (g_counts) and sort to get the lists indices to match
avt_pp = round(t_p_v.sort_values(ascending=False) / g_counts,2)
Gender_Purchasing_Analysis = pd.DataFrame(
                                        {'Gender':genders,
                                        'Purchase Count':p_cnt,
                                        'Average Purchase Price':av_p_p,
                                        'Total Purchase Value':t_p_v,
                                        "Avg Total Purchase per Person" : avt_pp})
Gender_Purchasing_Analysis = Gender_Purchasing_Analysis.set_index('Gender')
Gender_Purchasing_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins1 = [0,9,14,19,24,29,34,39,200]
labels1 = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','>40']
pymoli['Age Group'] = pd.cut(pymoli['Age'],bins1,labels = labels1)
dem2 = pymoli.groupby(['Age Group','SN'])
#the next line is magic /gibberish
# again, more complex and less efficient than is desirable but it works
dem2 = dem2.size().to_frame().reset_index()

num_ages = {}
for x in labels1:
    y = round(len(dem2.loc[dem2['Age Group'] == x]))
    num_ages.update({x:y})
num_prcnt = {}
for x1 in num_ages:
    y2 =round((num_ages[x1]/total_players)*100,2)
    num_prcnt.update({x1:y2})
Age_Analysis = pd.DataFrame([num_ages,num_prcnt])
Age_Analysis = Age_Analysis.rename(index={0:'Total Count',1:'Percent of Players'})
Age_Analysis = Age_Analysis[labels1]
Age_Analysis
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Starts getting normal
dem3 = pymoli.groupby(['Age Group'])
dem3.head()
age_price_count = dem3["Price"].count()
age_price_total = dem3["Price"].sum()
age_price_mean = round(dem3["Price"].mean(),2)
#get list of ages from dictionary in last section
list_ages = [num_ages[x] for x in num_ages]
age_price_avgper = age_price_total / list_ages
Gender_Purchasing_Analysis = pd.DataFrame(
                                        {'Purchase Count':age_price_count,
                                        'Average Purchase Price':age_price_mean,
                                        'Total Purchase Value':age_price_total,
                                        "Avg Total Purchase per Person" : age_price_avgper})
Gender_Purchasing_Analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
dem4 = pymoli.groupby(['SN'])
pur_user = dem4["Price"].count()#.sort_values(ascending=False)
t_pur_user = dem4["Price"].sum()#.sort_values(ascending=False)
av_pur_user = round(t_pur_user / pur_user,2)
User_Purchasing_Analysis = pd.DataFrame(
                                        {'Purchase Count':pur_user,
                                        'Average Purchase Price':av_pur_user,
                                        'Total Purchase Value':t_pur_user})
User_Purchasing_Analysis = User_Purchasing_Analysis.sort_values('Total Purchase Value', ascending=False)
User_Purchasing_Analysis.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = pymoli[['Item ID','Item Name','Price']]
ig = items.groupby(["Item ID","Item Name"])
item_pc = ig['Price'].count()#.sort_values(ascending=False)
item_pt = ig['Price'].sum()
item_pr = ig['Price'].mean()
Item_Purchasing_Analysis = pd.DataFrame(
                                        {'Purchase Count':item_pc,
                                        'Item Price':item_pr,
                                        'Total Purchase Value':item_pt})
Item_Purchasing_Analysis = Item_Purchasing_Analysis.sort_values('Purchase Count', ascending=False)
Item_Purchasing_Analysis.head()
```
